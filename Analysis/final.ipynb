```python
import pandas as pd

df = pd.read_csv("../dataset/symptoSense.csv")
```

```python
symptoms = df.drop('prognosis', axis=1).columns
print(symptoms)
```

```python
symptom_list = list(symptoms)
print(symptom_list)
```

```python
max_words = 0
that_word=""
for symptom in symptom_list:
    spaces = 1
    for ch in symptom:
        if ch == " ":
            spaces += 1
    if spaces > max_words:
        max_words = spaces
        that_word = symptom
print(max_words)
print(that_word)
```

```python
symptoms_with_or = []
for symptom in symptom_list:
    word = ""
    for ch in symptom:
        if ch == " ":
            if word == "or":
                symptoms_with_or.append(symptom)
            word = ""
        else:
            word+=ch
print(symptoms_with_or)     
```
